```python
# Import your libraries here (numpy)
import numpy as np
```

```python
# Define number of users as 5, number of movies as 15
users = 5
movies = 15
```

```python
# Generate a 15 x 5 or 5 x 15 numpy table filled with random numbers between 1.0 and 9.0. 
# These will represent movie ratings Fill these with randomly generated integers (each integer equally likely).
# (page 40 and 42, chapter 2). Used a fixed seed so the result is reproducable.

random_matrix = np.random.randint(1, 10, size=(movies, users))
print(random_matrix)
```

```python
# Divide all the numbers by 2, so that you work with half stars.
matrix = np.divide(random_matrix, 2)
print(matrix)
```

```python
# Compute the minimum score that each movie has received. How many movies have the minimum score?
occurences = 0
min = 0.5
for i in matrix:
    score = np.amin(i)
    if score == min:
        occurences += 1

print(occurences)
```

```python
# Compute the total score that each movie has received by adding the total score. Store these in a separate array.
total_scores = []
for i in matrix:
    total_scores.append(np.sum(i))

print(total_scores)
```

```python
# Show the total score for each movie in a histogram.
import matplotlib.pyplot as plt

np.histogram(total_scores)
_ = plt.hist(total_scores, bins='auto')
plt.title("Total Score for Each Movie")
plt.xlabel("score")
plt.ylabel("occurences")
plt.show()
```

```python
# Compute the average of the total scores for each movie, 
# and look at how far away the scores of each movie are away from the average. 
# Do you think that the total scores are uniformly distributed (each value equally likely)? 
# Of it it closer to a normal distribution (values tend to be fairly close to the average)?

average = np.average(total_scores)
print(average)
variation = total_scores - average
print(variation)

np.histogram(variation)
_ = plt.hist(variation, bins='auto')
plt.title("Variation of Average Scores")
plt.xlabel("variation")
plt.ylabel("occurences")
plt.show()
```

As we can see, the average scores tend towards a normal distribution, with the total score likely to be close to the average score

```python
# Compute the standard deviation for each movie. 
# Which movie is most divisive (highest standard deviation)? 
# Which movie is least divisive?

for i in matrix:
    std = np.std(i)
    print(f'Movie Set {i}. Standard Deviation = {std}')
```

This shows us that the highest deviation is 1.463 in the 3rd movie, while the lowest is 0.8 in the 6th movie.

```python
# On our movie website, we only display review texts from users that gave a score of 3 stars or higher. 
# Can you count for each movie how many reviews of 3 stars or more have been received?

count = np.count_nonzero(matrix >= 3, axis=1)
print(f'Scores of 3 or Greater {count}')
```

```python
# Use python code to check whether there a movie that has received 
# both a 4.5 or 5 star review and a one star or 1.5 star review.
count = 0
high_scores = np.count_nonzero(matrix >= 4.5, axis=1)
low_scores = np.count_nonzero(matrix <= 1.5, axis=1)
for i in range(len(high_scores)):
    if high_scores[i] >0 and low_scores[i] >0:
        count += 1
        print(f'There is a movie with scores of 4.5 or higher and scores of 1.5 or lower')
print(f"As we can see, there are {count} cases where a movie recieved both a very low and very high rating.")
```
